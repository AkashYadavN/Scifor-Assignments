```python
# Python program to demonstrate
# hybrid inheritance


class Karnataka:
    def func1(self):
        print("This function is an Area.")


class Mysore(Karnataka):
    def func2(self):
        print("This function is in Mysore. ")


class Bangalore(Karnataka):
    def func3(self):
        print("This function is in Bangalore.")


class Palace(Mysore, Karnataka):
    def func4(self):
        print("This function is in Mysore.")


# Driver's code
object = Palace()
object.func1()
object.func2()
```
